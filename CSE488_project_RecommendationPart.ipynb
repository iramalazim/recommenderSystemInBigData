##Reading Data

```python
DataFile = open('msnbc990928.seq')
readLines = DataFile.readlines()
colName = readLines[0]
```

```python
print(colName)
```

```python
colName = ['userId'] + colName.split()
colId = list(range(1,len(colName)))
colId
```

Original Sequence Data

```python
sequencesOfData = [[int(data) for data in Line.split()] for Line in readLines[1:]]
NumberOfRecords = len(sequencesOfData)
NumberOfRecords
```

```python
sequencesOfData
```

Converting the data into set for removing duplicacy

```python
import pandas as pd
dataIntoSet = [set(x) for x in sequencesOfData]
dataIntoSet
#print(type(dataIntoSet) )
```

```python

list_data = [];
user_id =0;
page_id = 0;
for seq in dataIntoSet:
  for page in seq:
    list_data.append([user_id,page])
    page_id+=1
  user_id+=1

list_data
```

```python
final_data = pd.DataFrame(list_data)
final_data.columns=["user_id","visited_pages"]
final_data.head(10)
```

```python
df3 = spark.createDataFrame(final_data)
df3.show()
```

```python
from pyspark.sql import functions as f
final_data = df3.groupBy("user_id").agg(f.collect_list("visited_pages").alias("visited_pages")).sort("user_id")
final_data.show()
```

##Part - II (Recommender systems)



FP Growth

```python
from pyspark.ml.fpm import FPGrowth
```

```python
fpGrowth = FPGrowth(itemsCol="visited_pages", minSupport=0.01, minConfidence=0.01)
model = fpGrowth.fit(final_data)
```

```python
model.freqItemsets.show()
```

```python
model.associationRules.show()
```

```python
rules = model.associationRules
frequentItemsets = model.freqItemsets
```

```python
rules.filter(rules.lift <1).show()
```

```python
frequentItemsets.filter( "freq>20000").show()
```

```python
transformed = model.transform(final_data)
transformed.show()
```

ALS Model

```python
from pyspark.ml.evaluation import RegressionEvaluator 
from pyspark.ml.recommendation import ALS 
from pyspark.ml.tuning import TrainValidationSplit
```

```python
model_data = df3.groupby("user_id","visited_pages").agg(count("visited_pages").alias("page_count")).sort(asc("user_id"))
model_data.show()
```

```python
(train_data,test_data) = model_data.randomSplit([0.8, 0.2])
```

```python
als = ALS(maxIter = 5, regParam = 0.01, userCol='user_id', itemCol='visited_pages', ratingCol='page_count',nonnegative=True)
model = als.fit(train_data)
predictions = model.transform(test_data)
```

```python
# predictions.show()
```

```python
no_null_pred = predictions.na.drop()
```

```python
no_null_pred.show()
```

```python
allUsers_rec = model.recommendForAllUsers(5)
allUsers_rec.show()
```

```python
evaluation = RegressionEvaluator(metricName='rmse',labelCol='page_count',predictionCol='prediction')
rmse = evaluation.evaluate(no_null_pred)
print("RMSE =",rmse)
```
